```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 적금

* 은행
* P2P
* 투자

```python
def get_pnl(name, savings_per_month, years, interest):
    months = years * 12
    interest = interest ** (1/12)

    df = pd.Series([interest]*months).to_frame(0)
    df.iloc[0, 0] = savings_per_month
    for i in range(months):
        if i > 0:
            df[i] = df[0].shift(i)
    return df.cumprod().sum(1).to_frame(name)
```

```python
savings_per_month = 100
years = 5

a = get_pnl('raw', savings_per_month, years, 1)
b = get_pnl('bank', savings_per_month, years, 1.01)
c = get_pnl('p2p', savings_per_month, years, 1.065)
d = get_pnl('10p', savings_per_month, years, 1.1)
e = get_pnl('20p', savings_per_month, years, 1.2)
f = get_pnl('50p', savings_per_month, years, 1.5)

join = pd.concat([a,b,c,d,e,f], 1)
join.plot(title='< after 5 years >')
plt.show()

print('%s 년동안 매달 %s 만원씩 투자' % (years, savings_per_month))
join.iloc[-1].to_frame('최종 누적')
```
